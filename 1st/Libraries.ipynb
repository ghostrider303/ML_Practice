```python
x=np.linspace(-2*np.pi,2*np.pi,num=100)
plt.plot(x)
plt.show()
```

```python
plt.figure(0, figsize=(10, 5))
plt.plot(np.sin(x), 'r')
plt.plot(np.cos(x), 'b')
plt.show()

plt.savefig('wave')
```
